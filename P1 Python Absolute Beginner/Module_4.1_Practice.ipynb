### Program: shirt order 
First get input for color and size  
- White has sizes L, M 
- Blue has sizes M, S  

print available or unavailable, then  
print the order confirmation of color and size  

*
**hint**: set a variable "available = False" before nested if statements and  
change to True if color and size are avaliable*

```python
# [ ] create shirt order using nested if 
# [ ] create shirt order using nested if

color = input("Would you like to order a white or blue shirt? ").lower()
size = input("What size shirt would you like? ").lower()

available = False

if color.startswith("w"):
    if size == "l" or size == "m":
        available = True
elif color.startswith("b"):
    if size == "m" or size == "s":
        available = True

if available:
    print("Available!")
    print("Order confirmed:", color, size)
else:
    print("Unavailable")
    print("Sorry, we don't carry that color/size combination.")
```

```python
# [ ] use a "forever" while loop to get user input of integers to add to sum, 
# until a non-digit is entered, then break the loop and print sum
int_num = input("Enter an integer number: ")
sum = 0

while int_num.isdigit():
    sum += int(int_num)
    int_num = input("Enter an integer number: ")

print(sum)
```

```python
# [ ] use a while True loop (forever loop) to give 4 chances for input of a correct color in a rainbow
# rainbow = "red orange yellow green blue indigo violet"
rainbow = "red orange yellow green blue indigo violet"
color_guess = input("Please guess a color from the rainbow: ")
if color_guess.lower() in rainbow:
    print("You choose the right color on the first try!")
else:
    guess2 = input("Nope, try again: ")
    if guess2.lower() in rainbow:
        print("You choose the right color on the second try!")
    else:
        guess3 = input("Nope, try again: ")
        if guess3.lower() in rainbow:
            print("You choose the right color on the third try!")
        else:
            guess4 = input("Nope, try again: ")
            if guess4.lower() in rainbow:
                print("You choose the right color on the fourth try!")
            else:
                print("Sorry, you're out of tries.")
```

### create a function: quiz_item()  that asks a question and tests if input is correct  
- quiz_item()has 2 parameter **strings**: question and solution  
- shows question, gets answer input  
- returns True if `answer == solution` or continues to ask question until correct answer is provided  
- use a while loop

create 2 or more quiz questions that call quiz_item()  
**Hint**: provide multiple choice or T/F answers

```python
# Create quiz_item() and 2 or more quiz questions that call quiz_item()
def quiz_item(question, solution):
    answer = input(question)
    while answer.lower() != solution.lower():
        print("That's not correct. Try again: ")
        answer = input(question)
    print("That's correct!")
    return True
quiz_item("What is the capital of France? (a) Marseille (b) Paris (c) Lyon (d) Nice: ", "b")
quiz_item('T/F: The "P" in PEMDAS stands for "Parentheses": ', "T")
```
