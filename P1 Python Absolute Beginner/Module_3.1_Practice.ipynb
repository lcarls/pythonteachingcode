### Guessing a letter A-Z  
**check_guess()** takes 2 string arguments: **letter and guess** (both expect single alphabetical character)   
    - if guess is not an alpha character print invalid and return False
    - test and print if guess is "high" or "low" and return False
    - test and print if guess is "correct" and return True

```python
# [ ] create check_guess()
def check_guess(letter,guess):
    guess = str(guess)
    if guess.isalpha():
        #guess is alphabetical letters
        if len(guess)==1:
            #guess is one character
            #TODO check the guess
            if guess == letter:
                print("Correct")
                return True
            elif guess > letter:
                print("Low")
                return False
            elif guess < letter:
                print("High")
                return False
            else:
                print("unhandled exception")
                return False
        else:
            print("Invalid - more than one character")
            return False
    else:
        print("Invalid - not alphabetical")
        return False
    
def letter_game(correct_answer):
    this_guess = input("Guess the letter (provide one letter): ")
    if True == check_guess(correct_answer,this_guess):
        print("On the first try! What a mensch.")
    else:
        #second guess
        this_guess = input("Guess the letter (provide one letter): ")
        if True == check_guess(correct_answer,this_guess):
            print("On the second try! What a mensch.")
        else:
        #third guess
            this_guess = input("Guess the letter (provide one letter): ")
            if True == check_guess(correct_answer,this_guess):
                print("On the third try! Sortta slow.")
            else:
                print("You lose.")

letter_game("A")
```

### Rainbow colors
ask for input of a favorite rainbow color first letter: ROYGBIV  

Using `if`, `elif`, and `else`:  
- print the color matching the letter  
    - R = Red  
    - O = Orange  
    - Y = Yellow  
    - G = Green
    - B = Blue
    - I = Indigo
    - V = Violet
    - else print "no match"


```python
# [ ] complete rainbow colors
rainbow_color = input("What is the first letter of your favorite rainbow color? ")

if rainbow_color == "R":
    print("Red")
elif rainbow_color == "O":
    print("Orange")
elif rainbow_color == "Y":
    print("Yellow")
elif rainbow_color == "G":
    print("Green")
elif rainbow_color == "B":
    print("Blue")
elif rainbow_color == "I":
    print("Indigo")
elif rainbow_color == "V":
    print("Violet")
else:
    print("no match")
```

```python
# [ ] make the code above into a function rainbow_color() that has a string parameter, 
# get input and call the function and return the matching color as a string or "no match" message.
# Call the function and print the return string.
def rainbow_color(color_letter):
    if color_letter == "R":
        return "Red"
    elif color_letter == "O":
        return "Orange"
    elif color_letter == "Y":
        return "Yellow"
    elif color_letter == "G":
        return "Green"
    elif color_letter == "B":
        return "Blue"
    elif color_letter == "I":
        return "Indigo"
    elif color_letter == "V":
        return "Violet"
    else:
        return "no match"
color_letter = input("What is the first letter of your favorite rainbow color? ")
result = rainbow_color(color_letter)
print(result)
```

**Create function age_20() that adds or subtracts 20 from your age for a return value based on current age** (use `if`) 
- call the funtion with user input and then use the return value in a sentence  
example `age_20(25)` returns **5**: 
> "5 years old, 20 years difference from now"

```python
# [ ] complete age_20()
def age_20(age):
    if age > 20:
        return age - 20
    else:
        return age + 20

input_age = int(input("Enter your current age: "))
result = age_20(input_age)
print(str(result) + " years old, 20 years difference from now")
```

**create a function rainbow_or_age that takes a string argument**
- if argument is a digit return the value of calling age_20() with the str value cast as **`int`** 
- if argument is an alphabetical character return the value of calling rainbow_color() with the str
- if neither return FALSE

```python
# [ ]  create rainbow_or_age()
def rainbow_or_age(input_value):
    if input_value.isdigit():
        return age_20(int(input_value))
    elif input_value.isalpha():
        return rainbow_color(input_value.upper())
    else:
        return False
input_value = input("Enter your age or the first letter of your favorite rainbow color: ")
result = rainbow_or_age(input_value)
print(result)
```
